```python
question = ["1. How many people have heart disease and how many people doesn't have heart disease?",
            "2. People of which sex has most heart disease?"
            "3. People of which sex has which type of chest pain most?"
            "4. People with which chest pain are most pron to have meart disease?",
            "5. What is the most common type of chest pain in the dataset?",
            "6. How many people have a blood pressure higher than 130?",
            "7. What is the average cholesterol level in the dataset?"]
question
```

```python
#LET'S find the answer of first question 
#1. How many people have heart disease and how many people doesn't have heart disease?

#GETTING the values 

df.target.value_counts()
```

```python
#PLOTING a pie chart 

df.target.value_counts().plot(kind = 'pie', figsize=(7,7))
plt.legend(["Disease", "No disease"]);
```

```python
#'0' represnt 'FEMALE'
#'1' represnt 'MALE'

#SEX column part

#'0' represnt 'NO disease'
#'1' represnt 'Disease'

#LET'S find the answer of our 2nd Question.
#2. People pf which sex has most heart disease?
pd.crosstab(df.target, df.sex)
```

```python
sns.countplot (x='target', data = df, hue = 'sex')

plt.title("HEART DISEASE FREQUENCY FOR SEX")
plt.xlabel ("0= No Heart Disease, 1 = Heart Disease");
```

```python
#3. People of which sex has which type of chest pain most?
#COUNTING values for different chest pain

df.cp.value_counts()
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind = 'bar', color = ['hotpink','lightblue','springgreen','yellow'])
plt.title('TYPE OF CHEST PAIN FOR SEX')
plt.xlabel ('0= Female, 1=Male')
plt.xticks(rotation=0);
```

```python
#4. People with which chest pain are most pron to have heart disease?
pd.crosstab(df.cp, df.target)
```

```python
sns.countplot(x='cp', data =df, hue='target');
```

```python
#MOST of people who has 'type 0' chest pain has less chance of heart disease.
#AND we see the opposite for other types.

#NOW let's take look at our age column

#CREATE a distribution plot with normal distribution curve

sns.displot ( x ='age', data = df, bins = 50, kde=True);
```

```python
# 5. What is the most common type of chest pain?
chest_pain_counts = df['cp'].value_counts()
print(f"The most common type of chest pain is type {chest_pain_counts.idxmax()}.")

# Plotting a bar chart for chest pain types
chest_pain_counts.plot(kind='bar', color='lightblue')
plt.title("CHEST PAIN TYPE DISTRIBUTION")
plt.xlabel("Chest Pain Type")
plt.ylabel("Frequency")
plt.xticks(rotation=0);

```

```python
# 6. How many people have a blood pressure higher than 130?"
# Counting the number of people with blood pressure higher than 130
high_bp_count = df[df['trestbps'] > 130].shape[0]
print(f"There are {high_bp_count} people with a blood pressure higher than 130.")

#  Plotting a histogram for resting blood pressure
df['trestbps'].plot(kind='hist', bins=20, color='orange')
plt.title("BLOOD PRESSURE DISTRIBUTION")
plt.xlabel("Resting Blood Pressure (trestbps)")
plt.ylabel("Frequency");
```

```python
# 7. What is the average cholesterol level in the dataset
average_chol = df['chol'].mean()
print(f"The average cholesterol level in the dataset is {average_chol:.2f} mg/dL.")

# Plotting a histogram for cholesterol levels
df['chol'].plot(kind='hist', bins=20, color='purple')
plt.title("CHOLESTEROL LEVEL DISTRIBUTION")
plt.xlabel("Cholesterol Level (mg/dL)")
plt.ylabel("Frequency");
```
